```python
import pandas as pd
import numpy as np
```

```python
excel_path = "data.xlsx"
df = pd.read_excel(excel_path, sheet_name="Sheet1")
```

```python
first_col = df.columns[0]
if np.issubdtype(df[first_col].dtype, np.datetime64) or "date" in str(first_col).lower():
    df = df.set_index(first_col)
```

```python
num_df = df.select_dtypes(include=[np.number]).copy()
if num_df.shape[1] == 0:
    num_df = df.apply(pd.to_numeric, errors="coerce")
```

```python
p95_abs = num_df.abs().quantile(0.95).median()
data_kind = "returns" if p95_abs < 0.5 else "prices"
print(f"Detected data type: {data_kind}")
```

```python
returns = num_df.pct_change().dropna(how="all")
```

```python
returns = returns.dropna(axis=1, how="all")
```

```python
output_csv = "cleaned_returns.csv"
returns.to_csv(output_csv, index=True)
```

```python
print(f"Saved cleaned returns to {output_csv} with shape {returns.shape}")
```
